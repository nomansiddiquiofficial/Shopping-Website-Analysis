```python
import os
import pandas as pd
```

#### Task: 1 =>  Merge 12 months of Sales Data into Single csv

```python

all_months_files = [file for file in os.listdir('./Sales_Data')]
```

```python
ourData = pd.DataFrame()

for a_month in all_months_files:
    df = pd.read_csv("./Sales_Data/"+a_month)  
    ourData = pd.concat([ourData,df])
ourData   
```

#### Task: 2 => Clean Up the Data

##### Drop rows of NAN

```python
nanData = ourData[ourData.isna().any(axis=1)]   #to see nan values
nanData.head(50)
```

```python
ourData = ourData.dropna(how = 'all')
ourData
```

##### find 'OR' in dataset and delete it

```python
ourData = ourData[ourData['Order Date'].str[0:2] != 'Or']
ourData
```

#### Task: 3 => Add a month Column

```python
ourData['Month'] = ourData['Order Date'].str[0:2]  
ourData['Month'] = ourData['Month'].astype('int32')                      #Convert String Month as interger
ourData
```

##### Convert str to numeric price and Quantity ordered

```python
ourData['Price Each'] = pd.to_numeric(ourData['Price Each'])
ourData['Quantity Ordered'] = pd.to_numeric(ourData['Quantity Ordered'])
```

#### Add a Sales Coulmn

```python
ourData['Sales'] = ourData['Price Each']*ourData['Quantity Ordered']
ourData
```

#### Task:4 => Question #1 What was the best month for Sales? How much was earned that month?


```python
results = ourData.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
```

```python
plt.bar(months,results['Sales'])
plt.xticks(months)

plt.xlabel('Months')
plt.ylabel('Prices in Million Dollars ($)')

plt.show()
```

#### Task: 5 =>  Question # 2 What City had the highest number of Sells?

##### Grab Only City  From Adress & Add a City Column 

```python
def getCity(address):
    return address.split(',')[1]                    #we can also do with funtions

# ourData['City']= ourData['Purchase Address'].apply(lambda x: getCity(x))  

"""As we hava a problem that in cty column we could have two same cities names of two differnt countries so we must include state 
names too"""

def getState(address):
    return address.split(',')[2].split(' ')[1]
ourData['City'] = ourData['Purchase Address'].apply(lambda x: getCity(x) + ' ' + getState(x)) 
# Add with city name
ourData
```

```python
cities_best_sales = ourData.groupby('City').sum()
cities_best_sales
```

```python
city = [city for city,ourData in ourData.groupby('City')]

plt.bar(city,cities_best_sales['Sales'])
plt.xticks(city, rotation='vertical',size= 10)

plt.xlabel('Cities, States')
plt.ylabel('Prices in Million Dollars ($)')

plt.show()
```

### Task: 7 => What products are most often sold together?

```python
df = ourData[ourData['Order ID'].duplicated(keep=False)]       
```

```python
df['Grouped Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df
```

```python
df = df[['Order ID','Grouped Products']].drop_duplicates()              #For Removing Duplicates Values 
df 
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped Products']:
    row_list = row.split(',')
    count.update(combinations(row_list,2))



for i,j in count.most_common():
    print(i,j)
    
    
    
    
    
    
    
    
```

```python
product_group = ourData.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]



prices = ourData.groupby('Product').mean()['Price Each']

fig, x = plt.subplots()

x1 = x.twinx()

x.bar(products,quantity_ordered,color = 'g')
x1.plot(products,prices)

x.set_xlabel("Products", color = 'b')
x.set_ylabel("Quantitty Ordered",color = 'b')
x1.set_ylabel("Prices", color = 'b')

x.set_xticklabels(products,rotation='vertical', size = 10)
plt.show()
```
